```python
import pandas as pd
import env
import numpy as np
from pydataset import data
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import acquire
import os
```

```python
def get_connection(db, user=env.user, host=env.host, password=env.password):
    return f'mysql+pymysql://{user}:{password}@{host}/{db}'

def get_attendance_data():
    filename = "tidy_data_attendence.csv"

    if os.path.isfile(filename):
        return pd.read_csv(filename)
    else:
        df = pd.read_sql('SELECT * FROM attendance', get_connection('tidy_data'))
        df.to_csv(filename)
        return df  
```

```python
attendance=get_attendance_data()
attendance
```

```python
attendance.rename(columns={'Unnamed: 0':"name"},inplace=True)
```

```python
attendance=attendance.replace(['P','T','H','A'],[1.0,0.9,0.5,0.0])
```

```python
tidy_attendance=attendance.melt(id_vars=['name'],value_name='grade')
tidy_attendance.groupby('name').grade.mean()
```

```python
def get_cake_data():
    filename = "tidy_data_cake.csv"

    if os.path.isfile(filename):
        return pd.read_csv(filename)
    else:
        df = pd.read_sql('SELECT * FROM cake_recipes', get_connection('tidy_data'))
        df.to_csv(filename)
        return df  
```

```python
cake=get_cake_data()
```

```python
cake_melt=cake.melt(id_vars='recipe:position',var_name='Temp',value_name='Score')
cake_melt
```

```python
cake_melt[['recipe', 'position']] = cake_melt['recipe:position'].str.split(':', expand = True)
cake_melt.drop(columns=['recipe:position'],inplace=True)
```

```python
cake_melt.groupby('recipe').Score.mean()
```

```python
cake_melt.groupby('Temp').Score.mean()
```

```python
cake_melt.sort_values(by=['Score'],ascending=False)
```
